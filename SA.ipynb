```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
#### Generate Metadata for each Image    

f = open('/Users/kahlil/Documents/GedinSaurS/~GGI/metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = "ADD_IMAGES_BASE_URI_HERE"
PROJECT_NAME = "GedinSaurS"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("background", i["background"]))
    token["attributes"].append(getAttribute("music", i["music"]))
    token["attributes"].append(getAttribute("species", i["species"]))
    token["attributes"].append(getAttribute("nature", i["nature"]))
    token["attributes"].append(getAttribute("# of attributes", i["# of attributes"]))
    token["attributes"].append(getAttribute("trait1", i["trait1"]))
    token["attributes"].append(getAttribute("trait2", i["trait2"]))
    token["attributes"].append(getAttribute("trait3", i["trait3"]))
    token["attributes"].append(getAttribute("trait4", i["trait4"]))
    token["attributes"].append(getAttribute("trait5", i["trait5"]))
    token["attributes"].append(getAttribute("trait6", i["trait6"]))
    token["attributes"].append(getAttribute("trait7", i["trait7"]))

    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
