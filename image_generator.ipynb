```python
## Generate Traits

TOTAL_IMAGES = 10000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #
    gearC = [];
    gearC_weights = [];
    
    #get basic attributes 
    new_image["background"] = random.choices(spag, spag_weights)[0]
    new_image["music"] = random.choices(music, music_weights)[0]
    new_image["species"] = random.choices(d_type, d_type_weights)[0]
    new_image["nature"] = random.choices(d_color, d_color_weights)[0]
    new_image["# of attributes"] = random.choices(numOfA, numOfA_weights)[0]
    
    ##assigining dino gear based on dino type
    if new_image["species"] == "Brontosaurus":
        gearC = gear1
        gearC_weights = gear1_weights
        
        
    if new_image["species"] == "Stegosaurus":
        gearC = gear2
        gearC_weights = gear2_weights
        
        
    if new_image["species"] == "Triceratops":
        gearC = gear3
        gearC_weights = gear3_weights
        
        
    if new_image["species"] == "Tyrannosaurus Rex":
        gearC = gear4
        gearC_weights = gear4_weights
        
        
    if new_image["species"] == "Pterosaur":
        gearC = gear5
        gearC_weights = gear5_weights
    
    
    ##assigining attributes based on number of attributes
    if new_image["# of attributes"] == 1:
        ## while loop to ensure traits are unique
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = "none2"
            new_image["trait3"] = "none3"
            new_image["trait4"] = "none4"
            new_image["trait5"] = "none5"
            new_image["trait6"] = "none6"
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
        
    if new_image["# of attributes"] == 2:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = "none3"
            new_image["trait4"] = "none4"
            new_image["trait5"] = "none5"
            new_image["trait6"] = "none6"
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
    
    if new_image["# of attributes"] == 3:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait4"] = "none4"
            new_image["trait5"] = "none5"
            new_image["trait6"] = "none6"
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
    
    if new_image["# of attributes"] == 4:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait4"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait5"] = "none5"
            new_image["trait6"] = "none6"
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
    
    if new_image["# of attributes"] == 5:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait4"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait5"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait6"] = "none6"
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
        
    if new_image["# of attributes"] == 6:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait4"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait5"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait6"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait7"] = "none7"
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
        
    if new_image["# of attributes"] == 7:
        while True:
            new_image["trait1"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait2"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait3"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait4"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait5"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait6"] = random.choices(gearC, gearC_weights)[0]
            new_image["trait7"] = random.choices(gearC, gearC_weights)[0]
            Alist = [new_image["trait1"], new_image["trait2"],new_image["trait3"], 
                 new_image["trait4"], new_image["trait5"],new_image["trait6"],
                new_image["trait7"]];
        
            if (len(set(Alist)) == len(Alist)):
                break
   
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```
